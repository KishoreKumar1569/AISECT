```python
#assigning the values for X (Independent values)
x = dataset[['Property_type', 'Location',  'Seller_type',
       'Area_sqft', 'Status', 'Security_deposit', 'Bathroom','BHK']]
```

```python
#assigning the vlaues for y (Dependent)
y = dataset['Rent_price']
```

```python
#showing the values(Data's of X)
x
```

```python
#showing the values of (data's of y)
y
```

```python
#importing train_test_split from the sklearn
from sklearn.model_selection import train_test_split
```

```python
#splitting the data X and Y
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
#importing DecisionTree and r2_Score from sklearn
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score
```

```python
#calling the DecisionTree function
dt = DecisionTreeRegressor()
dt.fit(x_train,y_train)
```

```python
#predicting the values using R2_score
y_pred = dt.predict(x_test)
dt_r2 = r2_score(y_test,y_pred)
```

```python
#gettin the accuracy
dt_r2
```
